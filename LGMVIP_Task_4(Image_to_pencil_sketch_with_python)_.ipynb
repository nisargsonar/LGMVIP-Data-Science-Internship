### Importing libraries


```python
import scipy
import cv2
import matplotlib.pyplot as plt
```

### Reading image file

```python
img = cv2.imread("parrot.jpg")
plt.imshow(img)
```

### With RGB

```python
color_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(color_img)
```

### Color to Grey image

```python
from IPython.display import display
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) 
plt.imshow(img_gray)
```

### Inverting grey image

```python
img_gray_inv = 255 - img_gray 
plt.imshow(img_gray_inv)
```

### Applying Gaussian Blue


```python
img_blur = cv2.GaussianBlur(img_gray_inv, ksize=(21, 21),sigmaX=0, sigmaY=0)
plt.imshow(img_blur)
```

### Pencil Sketch

```python
def dodgeV2(image, mask):
  return (cv2.divide(image, 160-mask, scale=162))

def burnV2(image, mask):
  return (255 - cv2.divide(255-image, 160-mask, scale=162))

#here we implemented blending to generate sketch
img_blend = dodgeV2(img_gray, img_blur)
```

```python
plt.imshow(img_blend)
```

```python
plt.imshow(color_img)
```

### Download image 

```python
cv2.imwrite('result.jpg',img_blend)
```
